```python
df = pd.read_csv("/kaggle/input/titanic/train.csv")
df_test = pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
#print(df.head())
print(df_test.head())
```

```python
df_test.info()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

```python

df_test["Fare"].fillna(df_test.groupby('Pclass')['Fare'].transform("median"),inplace=True)

df_test.drop(['Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

```python
df['Age']=df['Age'].fillna(df["Age"].mean())
print(df.isnull().sum())
df_test['Age']=df_test['Age'].fillna(df_test["Age"].mean())
print(df_test.isnull().sum())

print(df_test.info())
print(df_test.isnull().sum())
```

```python
print(df_test.info())
print(df.info())
print(df.isnull().sum())
print(df_test.isnull().sum())
```

```python
independent = df.drop(['Survived'],axis=1)
target = df['Survived']
```

```python
print(independent.head())
```

```python
independent['Fare'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
sex_le = LabelEncoder()
sex_test = LabelEncoder()
```

```python
independent['Sex_enc'] = sex_le.fit_transform(independent['Sex'])
independent.drop(['Sex'],axis=1,inplace=True)
```

```python
df_test['Sex_enc'] = sex_le.fit_transform(df_test['Sex'])
df_test.drop(['Sex'],axis=1,inplace=True)
```

```python
df_test.head()
```

**Modelling**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
#Cross VAlidation(K-fold)
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10,shuffle=True,random_state=0)
```

```python
#KNN
clf = KNeighborsClassifier(n_neighbors= 13)
scoring='accuracy'
score = cross_val_score(clf,independent,target,cv=k_fold,n_jobs=1,scoring = scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

```python
clf = DecisionTreeClassifier()
scoring='accuracy'
score = cross_val_score(clf,independent,target,cv=k_fold,n_jobs=1,scoring = scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```

```python
clf = RandomForestClassifier(n_estimators=13)
scoring='accuracy'
score = cross_val_score(clf,independent,target,cv=k_fold,n_jobs=1,scoring = scoring)
print(score)
round(np.mean(score)*100,2)
```

```python
clf = GaussianNB()
scoring='accuracy'
score = cross_val_score(clf,independent,target,cv=k_fold,n_jobs=1,scoring = scoring)
print(score)
round(np.mean(score)*100,2)
```

```python
clf = SVC()
scoring='accuracy'
score = cross_val_score(clf,independent,target,cv=k_fold,n_jobs=1,scoring = scoring)
print(score)
round(np.mean(score)*100,2)
```

***TESTING***

```python
clf = RandomForestClassifier(n_estimators=13)
clf.fit(independent,target)
```

```python
test = df_test.drop('PassengerId',1)
prediction = clf.predict(test)
```
